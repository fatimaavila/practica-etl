```python
import pandas as pd
import json
```

```python
data = json.load(open('mpd.slice.1000-1999.json'))
```

```python
playlists=data['playlists']
```

```python
df = pd.DataFrame.from_dict(playlists)
df.head(5)
```

```python
columnas=df['tracks'][1][0].keys()
#para obtener los datos de artista, album, 
#track se tiene que extraer del array que está dentro de las tracks de cada una,
#para luego pbtener cada uno.
columnas
```

```python
dffull = pd.DataFrame(columns=columnas)
dffull
```

```python
#se trabajrea con 5 playlist por ahora
# df=df.head(5)
df
```

```python
##borrando 
dffull = pd.DataFrame(columns=columnas)
dffull
```

```python
import warnings
warnings.filterwarnings('ignore')
## https://softhints.com/turn-off-warnings-jupyterlab-jupyter-notebook/
```

```python
for index, row in df.iterrows():
    #print(row['name'], row['pid'],row['num_tracks'])
    for row2 in row['tracks']:
      dftemp = pd.DataFrame(row2,index=[0])
      dffull = dffull.append(dftemp)
```

```python
dffull # se tienen todas las canciones juntas
```

```python
uartist = dffull['artist_uri'].unique()
## uartist
```

```python
dffull[dffull["artist_uri"]=='spotify:artist:3TymzPhJTMyupk7P5xkahM'].iloc[0]
```

```python
##crear un df de artist vacio
dfartist = pd.DataFrame(columns=["sk_id","bk_id","name"]) 

rownum=1
for row in uartist:
  ##print(dffull[dffull.artist_uri == row].iloc[0]["artist_uri"],dffull[dffull.artist_uri == row].iloc[0]['artist_name'])

##https://vitalflux.com/pandas-dataframe-how-to-add-rows-columns/#:~:text=There%20are%20multiple%20ways%20of,use%20Dataframe.insert()%20method.
  dftemp = pd.DataFrame({
      "sk_id":rownum,
      "bk_id":dffull[dffull.artist_uri == row].iloc[0]["artist_uri"],
      "name":dffull[dffull.artist_uri == row].iloc[0]["artist_name"]
      },index=[0])
  dfartist=dfartist.append(dftemp)

  rownum = rownum + 1
```

```python
dfartist
```

```python
## lo mismo para albums
ualbum = dffull['album_uri'].unique()
# ualbum
```

```python
##crear un df de album vacio
dfalbum = pd.DataFrame(columns=["sk_id","bk_id","name"]) 

rownum=1
for row in ualbum:
  dftemp = pd.DataFrame({
      "sk_id":rownum,
      "bk_id":dffull[dffull.album_uri == row].iloc[0]["album_uri"],
      "name":dffull[dffull.album_uri == row].iloc[0]["album_name"]
      },index=[0])
  dfalbum=dfalbum.append(dftemp)

  rownum = rownum + 1
```

```python
dfalbum
```

```python
## lo mismo para tracks
utrack = dffull['track_uri'].unique()
utrack
```

```python
##crear un df de track vacio
dftrack = pd.DataFrame(columns=["sk_id","bk_id","name"]) 

rownum=1
for row in utrack:
  dftemp = pd.DataFrame({
      "sk_id":rownum,
      "bk_id":dffull[dffull.track_uri == row].iloc[0]["track_uri"],
      "name":dffull[dffull.track_uri == row].iloc[0]["track_name"]
      },index=[0])
  dftrack=dftrack.append(dftemp)

  rownum = rownum + 1
```

```python
dftrack
```

```python
##todo inserta a base de datos postgresql
## 1. d_albums desde dfalbum
## 2. d_artists desde dfartist
## 3. d_tracks  desde dftrack
## 4 insertar a d_playlists desde df
## 5 insertar a d_playlist_Tracks desde dffull
```

```python
## exportar datos transformados a csv como plan B para agregar a Postgres
dfalbum.to_csv("albums.csv",index=False)
dfartist.to_csv("artists.csv",index=False)
dftrack.to_csv("tracks.csv",index=False)
df.to_csv("playlists.csv",index=False)
dffull.to_csv("playlist_tracks.csv",index=False)
```
